```python
from matplotlib import pyplot as plt
from pathlib    import Path
```

```python
path = Path('/Users/baranserajelahi/Codes/fraud-detection-pytorch-scikit-fastai/data')
Path.BASE_PATH = path
```

```python
NN  = plt.imread(path.BASE_PATH/'../figures/NN_cm.jpg')
RUS = plt.imread(path.BASE_PATH/'../figures/RUS_NN_cm.jpg')
ROS = plt.imread(path.BASE_PATH/'../figures/ROS_NN_cm.jpg')
CW  = plt.imread(path.BASE_PATH/'../figures/CW_NN_cm.jpg')

images = [NN, RUS, ROS, CW]
titles = ['no imbalance correction','random under-sampling','random over-sampling','class weighting']
labels = ['A','B','C','D']
aurocs = [0.82, 0.8, 0.82, 0.80]
```

```python
fig, ax = plt.subplots(2,2, figsize=(12,12))
ax = ax.flatten()
fig.suptitle('Neural Network', fontsize=30)

for i,axis in enumerate(ax):
    axis.imshow(images[i])
    axis.axis('off')
    axis.set_title(titles[i], y=1.025, fontsize=20)
    axis.text(0.5, 0.5, f'{labels[i]}', fontsize=15)
    axis.text(0, 0, f'auroc = {aurocs[i]}', fontsize=15, transform=axis.transAxes,
              bbox=dict(facecolor='red', alpha=0.5))
    
fig.tight_layout()
fig.savefig(path.BASE_PATH/'../figures/NNs.jpg', dpi=1000, bbox_inches='tight')
```
